```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
history = pickle.load( open( "/opt/datos/entrenamientos/larraondo/trainHistoryDict_vgg16_0-2-6", "rb" ) )

print(f"val_loss: {history['val_loss'][-1]}")
print(f"loss: {history['loss'][-1]}")
#print(history)
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('VGG16 network training 1000-800-400 hPa')
plt.ylabel('mean absolute error')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```
