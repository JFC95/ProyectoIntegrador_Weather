```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.load('/opt/datos/dataset/sin_normalizar/X.npy')
X.shape
```

```python
X_0 = X[:, 0, :, :]
X_1 = X[:, 1, :, :]
X_2 = X[:, 2, :, :]

print(X_0.shape)
print(X_1.shape)
print(X_2.shape)
```

```python
X_0 = X_0.flatten()
X_1 = X_1.flatten()
X_2 = X_2.flatten()
```

```python
intervalos = range(int(min(X_0)), int(max(X_0))) #extremos de los intervalos
plt.hist(x=X_0, bins=intervalos, color='#F2AB6D', rwidth=0.85)
plt.xticks([int(X_0.min()), int(X_0.mean()), int(X_0.max())]) #para X_0

plt.title('1000 hPa')
plt.show() #dibujamos el histograma
```

```python
intervalos = range(int(min(X_1)), int(max(X_1))) #extremos de los intervalos
plt.hist(x=X_1, bins=intervalos, color='#F2AB6D', rwidth=0.85)
plt.xticks([int(X_1.min()), int(X_1.mean()), int(X_1.max())]) #para X_1

plt.title('700 hPa')
plt.show() #dibujamos el histograma
```

```python
intervalos = range(int(min(X_2)), int(max(X_2))) #extremos de los intervalos
plt.hist(x=X_2, bins=intervalos, color='#F2AB6D', rwidth=0.85)
plt.xticks([int(X_2.min()), int(X_2.mean()), int(X_2.max())]) #para X_2

plt.title('500 hPa')
plt.show() #dibujamos el histograma
```
