```python
import os
import pandas as pd
from pathlib import Path
import georasters as gr
import datetime
import numpy as np
from pathlib import Path
import re
import tensorflow as tf
```

```python
directory_in_str = "/opt/datos/gfsRaster/matrices_normalizadas/" #dentro de ....
pathlist = Path(directory_in_str).glob('**/*.csv')

lista_de_path = list()
lista_dates = list()
i=0
j=0

np_image = np.empty((3,176,176))

start_date = "2020-01-01 12:00"
finish_date = "2021-01-01 09:00"
idx = pd.date_range(start_date, finish_date, freq='3H').strftime("%Y-%m-%d %H:%M")

for path in pathlist:
    path_in_str = str(path)
    lista_de_path.append(path_in_str)
    
for time in idx:
    time_new = time.replace(" ", "Z")
    lista_dates.append(time_new)

cant = len(lista_de_path) // 3
np_all_images = np.empty((cant,3,176,176))
    
for times in lista_dates:
    for file in lista_de_path:
        if(times in file):
            df = pd.read_csv(file)
            np_matriz = df.to_numpy() # 176x176
            #print("%d %s" %(i,file))
            np_image[i] = np_matriz
            i=i+1
            if(i==3):
                np_all_images[j] = np_image
                #print("--------------------------------------------------------------------------------")
                j=j+1
                i=0
```

```python
print(np_image.shape)
print(np_all_images.shape)
```

```python
np_Xtrain = np_all_images[0:2050]
np_Xval = np_all_images[2050:2489]
np_Xtest = np_all_images[2489:2928]
print(np_all_images.shape)
print(np_Xtrain.shape)
print(np_Xval.shape)
print(np_Xtest.shape)
```

```python
np.save("X_train.npy",np_Xtrain)
np.save("X_val.npy",np_Xval)
np.save("X_test.npy",np_Xtest)
```
