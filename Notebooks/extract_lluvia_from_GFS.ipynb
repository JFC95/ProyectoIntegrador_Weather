```python
def getPpnBand(grib):
    """ De un Dataset de gdal devuelve la banda donde se encuentra
    la variable APCP
    """
    dictVar = {}
    for band in range(1, grib.RasterCount + 1):
        var = grib.GetRasterBand(band)
        if var.GetMetadata()['GRIB_ELEMENT'] in ("ACPCP", "APCP06"):
            dictVar[int(band)] = "aca"

    return dictVar

getPpnBand(grib)
```
