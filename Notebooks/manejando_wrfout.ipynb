```python
import xarray as xr
from netCDF4 import Dataset
import wrf
```

```python
# wrfout son los archivos de salida de wrf, cuya extension es .nc
```

```python
# Con xarray
filename = '/opt/wrfdatos/wrfout/2020_01/wrfout_A_d01_2020-01-01_06:00:00'
ds = xr.open_dataset(filename)
```

```python
arrs = {}
for t in range(len(ds.variables['Times'])):
    arrs[t] = (ds.variables['RAINNC'].values[t, :, :] + ds.variables['RAINC'].values[t, :, :])
print(arrs)
```

```python
# Con Netcdf
ncwrf = Dataset(filename)
#print(dataset)
```

```python
variable1 = 'RAINC'
variable2 = 'RAINNC'
var1 = wrf.getvar(ncwrf, variable1, timeidx=wrf.ALL_TIMES)
var2 = wrf.getvar(ncwrf, variable2, timeidx=wrf.ALL_TIMES)
print(type(var1))
print(type(var2))
print(var1.shape)
print(var2.shape)
```
