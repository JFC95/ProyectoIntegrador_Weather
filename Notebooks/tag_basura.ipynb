```python
import os
from pathlib import Path
```

```python
directory_in_str = "/opt/datos/gfsRaster/geotiff/" #dentro de ....
pathlist = Path(directory_in_str).glob('**/*.tiff') 
horarios_a_borrar = ["Z12:00", "Z15:00", "Z18:00", "Z21:00"]
```

```python
for path in pathlist:
    path_in_str = str(path)
    if("HGT_100000_" not in path_in_str): #para tiff de 50k o 70k
        if("_06/" in path_in_str): #corrida de las 6
            if(("Z06:00" in path_in_str) or ("Z09:00" in path_in_str)): #6hs basura
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(path_in_str[37:39] != path_in_str[70:72]): #son de distinto dia entonces borro
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
        if("_18/" in path_in_str): #corrida de las 18
            if(path_in_str[37:39] == path_in_str[70:72]): #son de igual dia entonces borro (6hs basura)
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(int(path_in_str[70:72]) - int(path_in_str[37:39]) == 1): #es el dia despues
                if(path_in_str[72:78] in horarios_a_borrar):
                    try:
                        path_nuevo = path_in_str.replace("None", "basura")
                        os.rename(path_in_str, path_nuevo)
                        #os.remove(path_in_str)
                    except:
                        print("No se encontro el archivo")
            else: #seria pasado mañana y mas
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
    else: #para los tiff 100k ISBL
        if("_06/" in path_in_str): #corrida de las 6
            if(("Z06:00" in path_in_str) or ("Z09:00" in path_in_str)): #6hs basura
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(path_in_str[37:39] != path_in_str[71:73]): #son de distinto dia entonces borro
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
        if("_18/" in path_in_str): #corrida de las 18
            if(path_in_str[37:39] == path_in_str[71:73]): #son de igual dia entonces borro (6hs basura)
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(int(path_in_str[71:73]) - int(path_in_str[37:39]) == 1): #es el dia despues
                if(path_in_str[73:79] in horarios_a_borrar):
                    try:
                        path_nuevo = path_in_str.replace("None", "basura")
                        os.rename(path_in_str, path_nuevo)
                        #os.remove(path_in_str)
                    except:
                        print("No se encontro el archivo")
            else: #seria pasado mañana y mas
                try:
                    path_nuevo = path_in_str.replace("None", "basura")
                    os.rename(path_in_str, path_nuevo)
                    #os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
    
```
