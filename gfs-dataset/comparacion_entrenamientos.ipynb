```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
def comparar(filename1, legend1, filename2, legend2):
    history1 = pickle.load( open( filename1, "rb" ) )
    history2 = pickle.load( open( filename2, "rb" ) )

    #print(f"val_loss VGG16: {history1['val_loss'][-1]}")
    #print(f"val_loss Unet: {history2['val_loss'][-1]}")
    #print(history)
    plt.plot(history1['val_loss'])
    plt.plot(history2['val_loss'])
    plt.title('Validation MAE during training')
    plt.ylabel('mean absolute error')
    plt.xlabel('epoch')
    plt.legend([legend1, legend2], loc='upper right')
    plt.show()
```

# 400-800-1000 hPa  vs.  500-700-1000 hPa
## Ambos: channel_last - con shuffle

```python
filename1 = '/opt/datos/entrenamientos/gfs/05'
legend1   = '400-800-1000 hPa'
filename2 = '/opt/datos/entrenamientos/gfs/06'
legend2   = '500-700-1000 hPa'
comparar(filename1, legend1, filename2, legend2)
```

# con shuffle  vs.  sin shuffle
## Ambos: 400-800-1000 hPa  -  channel_last

```python
filename1 = '/opt/datos/entrenamientos/gfs/03'
legend1   = 'con shuffle'
filename2 = '/opt/datos/entrenamientos/gfs/04'
legend2   = 'sin shuffle'
comparar(filename1, legend1, filename2, legend2)
```

# Channel_first  vs.  Channel_last
## Ambos: 400-800-1000 hPa

```python
filename1 = '/opt/datos/entrenamientos/gfs/03'
legend1   = 'channel first'
filename2 = '/opt/datos/entrenamientos/gfs/04'
legend2   = 'channel last'
comparar(filename1, legend1, filename2, legend2)
```
