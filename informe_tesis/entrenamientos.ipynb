```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

# VGG16 Cordoba - HGT 500-700-1000

```python
history = pickle.load( open( "/opt/datos/entrenamientos/cba/01/hist", "rb" ) )

plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Enfoque 1: VGG16 network training 1000-700-500 hPa')
plt.ylabel('mean absolute error')
plt.xlabel('epoch')
plt.ylim(0.198, 0.24)
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```
