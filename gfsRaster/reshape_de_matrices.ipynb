```python
directory_in_str = "geotiff/2020_01/" #dentro de ....
pathlist = Path(directory_in_str).glob('**/*HGT_50000*.tiff') #todos los tiff de 50000 ISBL

cba_lon_min = -67
cba_lon_max = -61
cba_lat_min = -35.3
cba_lat_max = -29

for path in pathlist:
    path_in_str = str(path)
    #print(path_in_str)
    hgt50k = gr.from_file(path_in_str)
    df_hgt50k = hgt50k.to_pandas()
    

    #filtro por longitudes mayores a -67
    df_mask = (df_hgt50k.x>=cba_lon_min)
    filtered_long_min = df_hgt50k[df_mask]
    #print(filtered_df)

    #filtro por longitudes menores a -61
    df_mask = (filtered_long_min.x<=cba_lon_max)
    filtered_long = filtered_long_min[df_mask]
    #print(filtered_long)

    #filtro por latitudes mayores a -35.3
    df_mask = (filtered_long.y>=cba_lat_min)
    filtered_lat_min= filtered_long[df_mask]
    #print(filtered_lat_min)

    #filtro por latitudes menores a -29
    df_mask = (filtered_lat_min.y<=cba_lat_max)
    filtered_lat = filtered_lat_min[df_mask]
    
    df_hgt50k = filtered_lat
    
    # Me quedo solo con la columna value
    df_hgt50k = df_hgt50k["value"]
    
    # Reshapeo matriz
    np_hgt50k = df_hgt50k.to_numpy()
    np_hgt50k = np_hgt50k.reshape((701,668)) #cambiar esas dimensiones
    
    df_hgt50k = pd.DataFrame(np_hgt50k)
    
    #Guardo la matriz en un csv
    regex = "^geotiff\/\w{7}\/\w{5}\/\w{8}"
    regex2 = "\.tiff"
    filename = re.sub(regex, "matrices/matriz_input", path_in_str)
    filename = re.sub(regex2, ".csv", filename)
    df_hgt50k.to_csv(filename, index=False)
    #print(filename)
```
