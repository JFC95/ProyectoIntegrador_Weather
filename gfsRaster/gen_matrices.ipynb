```python
import os
import pandas as pd
from pathlib import Path
import georasters as gr
import datetime
import numpy as np
from pathlib import Path
import re
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
directory_in_str = "/opt/datos/gfsRaster/geotiff/2020_01/" #dentro de ....
pathlist = Path(directory_in_str).glob('**/*None_HGT_50000*.tiff') #todos los tiff de 50000 ISBL

for path in pathlist:
    path_in_str = str(path)
    #print(path_in_str)
    hgt50k = gr.from_file(path_in_str)
    df_hgt50k = hgt50k.to_pandas()
    
    # Me quedo solo con la columna value
    df_hgt50k = df_hgt50k["value"]
    
    # Reshapeo matriz
    np_hgt50k = df_hgt50k.to_numpy()
    np_hgt50k = np_hgt50k.reshape((175,175)) 
    
    df_hgt50k = pd.DataFrame(np_hgt50k)
    
    #Guardo la matriz en un csv
    regex = "geotiff\/\w{7}\/\w{5}\/\w{8}"
    regex2 = "\.tiff"
    filename = re.sub(regex, "matrices/matriz_input", path_in_str)
    filename = re.sub(regex2, ".csv", filename)
    df_hgt50k.to_csv(filename, index=False)
    #print(filename)
```
