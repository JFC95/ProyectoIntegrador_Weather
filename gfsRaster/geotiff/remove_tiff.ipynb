```python
from pathlib import Path
directory_in_str = "./"
#directory_in_str = "/home/joaquinfunes/ProyectoIntegrador_Weather/gfsRaster/geotiff"
pathlist = Path(directory_in_str).glob('**/*.tiff')
horarios_a_borrar = ["Z12:00", "Z15:00", "Z18:00", "Z21:00"]

for path in pathlist:
    path_in_str = str(path)
    if("HGT_100000_" not in path_in_str): #para tiff de 50k o 70k
        if("_06/" in path_in_str): #corrida de las 6
            if(("Z06:00" in path_in_str) or ("Z09:00" in path_in_str)): #6hs basura
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(path_in_str[8:10] != path_in_str[41:43]): #son de distinto dia entonces borro
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
        if("_18/" in path_in_str): #corrida de las 18
            if(path_in_str[8:10] == path_in_str[41:43]): #son de igual dia entonces borro (6hs basura)
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(int(path_in_str[41:43]) - int(path_in_str[8:10]) == 1): #es el dia despues
                if(path_in_str[43:49] in horarios_a_borrar):
                    try:
                        os.remove(path_in_str)
                    except:
                        print("No se encontro el archivo")
            else: #seria pasado mañana y mas
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
    else: #para los tiff 100k ISBL
        if("_06/" in path_in_str): #corrida de las 6
            if(("Z06:00" in path_in_str) or ("Z09:00" in path_in_str)): #6hs basura
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(path_in_str[8:10] != path_in_str[42:44]): #son de distinto dia entonces borro
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
        if("_18/" in path_in_str): #corrida de las 18
            if(path_in_str[8:10] == path_in_str[42:44]): #son de igual dia entonces borro (6hs basura)
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
            if(int(path_in_str[42:44]) - int(path_in_str[8:10]) == 1): #es el dia despues
                if(path_in_str[44:50] in horarios_a_borrar):
                    try:
                        os.remove(path_in_str)
                    except:
                        print("No se encontro el archivo")
            else: #seria pasado mañana y mas
                try:
                    os.remove(path_in_str)
                except:
                    print("No se encontro el archivo")
```
