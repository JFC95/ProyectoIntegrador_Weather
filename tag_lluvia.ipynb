```python
import os
import datetime
import pandas as pd
import csv
import re

x = 0
df2 = pd.DataFrame(columns=['fecha', 'LLUVIA / NO LLUVIA'])
path_datos = "wrf-cuenca/src/temp/tablas/"  
time_init = 6                    # Se ignoran las primeras 6 horas de cada archivo.
times = 12                       # Se toman 12 horas de cada archivo. (SON 2 ARCHIVOS POR DIA)
horas = ["06","18"]

index_horas = 1                  # Se empieza desde las 18hs del primer dia, para coordinar con las iteraciones siguientes
base = datetime.date(2020,1,1)  # Primer dia del dataset (SIEMPRE LOS DATOS VAN A INICIAR DESDE EL DIA SIGUIENTE A ESTE DIA INICIAL, DEBIDO
                                 #                         A QUE SE EMPIEZA DESDE LAS 18hs. Y SE IGNORAN LAS PRIMERAS 6hs.)
numdays = 2000                      # Cant. total de dias del dataset
date_list = [base + datetime.timedelta(days=x) for x in range(1, numdays)]  # Genera una lista de fechas con intervalo de un dia

# Se arma otro arreglo para obtener la lista de fechas en el formato del nombre del archivo del modelo (aaaa-mm-dd)
dias = []
paths = []
for x in range (0,numdays-1):
    dias.append(date_list[x].strftime("%Y-%m-%d"))

dataDIR = path_datos + "{}/{}/cordoba/cuencas/ppn_por_hora_A_CBA_A_{}.csv".format(base.strftime("%Y_%m"),base.strftime("%d"),horas[index_horas])
paths.append(dataDIR)
index_horas ^= 1                 # Se swapea para abrir el proximo archivo del mismo dia con la otra hora (6hs)        
 
for dia in dias:                                                           # Lee el resto de los dias del dataset
    for hora in horas:                                                     # Lee ambas horas de cada dia (6hs y 18hs)
        year = dia.split("-")[0]                                           # Obtiene anio y mes dia
        month = dia.split("-")[1]                                       
        day = dia.split("-")[2]
        dataDIR = path_datos + "{}_{}/{}/cordoba/cuencas/ppn_por_hora_A_CBA_A_{}.csv".format(base.strftime(year),base.strftime(month), base.strftime(day),hora)
        paths.append(dataDIR)
for p in range(0,len(paths)):
    try:
        df = pd.read_csv (paths[p])
        data_found = 1
        cuenca=df[9:21][['04-Cuenca R. Anizacate','fecha']]
        csv_stack = pd.DataFrame(columns=['fecha', 'LLUVIA / NO LLUVIA'])
        for i in range(0, 12):
            if(cuenca.iloc[i]['04-Cuenca R. Anizacate'] > 1.0):
                df2.loc[i+(x)] = [cuenca.iloc[i]['fecha'], 'SI LLUVIA']
                print("%s LLUVIA" % cuenca.iloc[i]['fecha'])
            else:
                df2.loc[i+(x)] = [cuenca.iloc[i]['fecha'], 'NO LLUVIA']
                print("%s NO LLUVIA" % cuenca.iloc[i]['fecha'])
            x=x+i
    except IOError as e:
        #print (e.filename)
        print("No encontrado %s" %e.filename)    

csv = df2.to_csv(r'04-Cuenca R. Anizacate.csv', index=False)             
```
