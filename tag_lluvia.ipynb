```python
import pandas as pd
```

```python
path = "2019_01/03/ppn_por_hora_CBA_A_18.csv"
df = pd.read_csv (path)
df
```

```python
df.shape
```

```python
a=df[9:21][['04-Cuenca R. Anizacate','fecha']]
a
```

```python
df2 = pd.DataFrame(columns=['fecha', 'LLUVIA / NO LLUVIA'])
df2.shape
```

```python
for i in range(0, 12):
    if(a.iloc[i]['04-Cuenca R. Anizacate'] > 0.2):
        #c = [(a.iloc[i]['fecha'], a.iloc[i]['04-Cuenca R. Anizacate'])]
        df2.loc[i] = [a.iloc[i]['fecha'], 'SI LLUVIA']
        print("%s LLUVIA" % a.iloc[i]['fecha'])
    else:
        df2.loc[i] = [a.iloc[i]['fecha'], 'NO LLUVIA']
        print("%s NO LLUVIA" % a.iloc[i]['fecha'])
```
