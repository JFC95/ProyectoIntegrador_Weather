```python
from tensorflow.keras.models import load_model
import numpy as np
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def graficar_entrenamiento(filename):
    history = pickle.load( open( filename, "rb" ) )

    print(f"val_loss: {history['val_loss'][-1]}")
    print(f"loss: {history['loss'][-1]}")
    #print(history)
    plt.plot(history['loss'])
    plt.plot(history['val_loss'])
    plt.title('Unet network training')
    plt.ylabel('mean absolute error')
    plt.xlabel('epoch')
    #plt.ylim(0,0.4)
    plt.legend(['train', 'validation'], loc='upper right')
    plt.show()
```

## Comparacion

```python
def comparar(filename1, legend1, filename2, legend2, filename3, legend3, filename4, legend4):
    history1 = pickle.load( open( filename1, "rb" ) )
    history2 = pickle.load( open( filename2, "rb" ) )
    history3 = pickle.load( open( filename3, "rb" ) )
    history4 = pickle.load( open( filename4, "rb" ) )

    #print(f"val_loss VGG16: {history1['val_loss'][-1]}")
    #print(f"val_loss Unet: {history2['val_loss'][-1]}")
    #print(history)
    plt.plot(history1['val_loss'])
    plt.plot(history2['val_loss'])
    plt.plot(history3['val_loss'])
    plt.plot(history4['val_loss'])
    plt.title('Validation MAE during training')
    plt.ylabel('mean absolute error')
    plt.xlabel('epoch')
    plt.legend([legend1, legend2, legend3, legend4], loc='upper right')
    plt.show()
```

```python
filename1, legend1 = '/opt/datos/entrenamientos/gfs/wind/05/history', '700 mb sin shuffle'
filename2, legend2 = '/opt/datos/entrenamientos/gfs/wind/06/history', '850 mb sin shuffle'
filename3, legend3 = '/opt/datos/entrenamientos/gfs/wind/07/history', '700 mb con shuffle'
filename4, legend4 = '/opt/datos/entrenamientos/gfs/wind/08/history', '850 mb con shuffle'
comparar(filename1, legend1, filename2, legend2, filename3, legend3, filename4, legend4)
```

## HGT vs Wind

```python
def comp(filename1, legend1, filename2, legend2):
    history1 = pickle.load( open( filename1, "rb" ) )
    history2 = pickle.load( open( filename2, "rb" ) )

    #print(f"val_loss VGG16: {history1['val_loss'][-1]}")
    #print(f"val_loss Unet: {history2['val_loss'][-1]}")
    #print(history)
    plt.plot(history1['val_loss'])
    plt.plot(history2['val_loss'])
    plt.title('Validation MAE Wind vs. HGT')
    plt.ylabel('mean absolute error')
    plt.xlabel('epoch')
    plt.legend([legend1, legend2], loc='upper right')
    plt.show()
```

```python
filename1, legend1 = '/opt/datos/entrenamientos/gfs/05', '400-800-1000 hPa con shuffle'
filename2, legend2 = '/opt/datos/entrenamientos/gfs/wind/06/history', '850 mb sin shuffle'
comp(filename1, legend1, filename2, legend2)
```
